```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv.imread('1QwjjwxGzQY.jpg')
image = cv.cvtColor(image1, cv.COLOR_BGR2GRAY)
```

```python
threshold = 115

# Примените пороговое значение
ret, binary_image = cv.threshold(image, threshold, 255, cv.THRESH_BINARY)

# Создайте ядро для эрозии
kernel = np.ones((3,3),np.uint8)

# Примените эрозию
erosion = cv.erode(binary_image, kernel, iterations = 1)

# Отобразите исходное и преобразованное изображения
plt.subplot(121),plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(erosion, cmap='gray')
plt.title('Eroded Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
cv.imwrite('binary_image.jpg', binary_image)
cv.imwrite('erosion.jpg', erosion)
```
