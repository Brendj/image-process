```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def equalize_image(image_path):
    # Загрузим изображение
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Применим эквализацию гистограммы
    equalized_image = cv2.equalizeHist(image)
    
    return equalized_image

def apply_negative(image):
    # Применим негатив
    negative_image = 255 - image
    
    return negative_image
```

```python
image_path = 'croped_image.jpg'
    
# Применяем эквализацию к фотографии
equalized_image = equalize_image(image_path)
    
# Применяем негатив к эквализированной фотографии
negative_image = apply_negative(equalized_image)
    
# Отобразим исходное изображение, эквализированное изображение и негатив
plt.figure(figsize=(12, 4))
    
plt.subplot(131)
plt.title('Исходное изображение')
plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
    
plt.subplot(132)
plt.title('Эквализированное изображение')
plt.imshow(equalized_image, cmap='gray')
    
plt.figure(figsize=(6, 6))  # Размер окна
plt.title('Негатив эквализированного изображения')
plt.imshow(negative_image, cmap='gray')
plt.show()
    
plt.show()
```
