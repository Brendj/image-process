```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def restore_image(image_path):
    # Загрузка изображения
    image = cv2.imread(image_path)
    
    # Преобразование изображения в оттенки серого
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Применение гауссового размытия для сглаживания полос
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    
    # Применение адаптивной пороговой обработки для получения бинарного изображения
    _, binary_image = cv2.threshold(blurred_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    # Нахождение и заполнение контуров для удаления полос
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 255, 255), -1)
    
    # Вывод восстановленного изображения
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Пример использования
image_path = 'path/to/your/image.jpg'
restore_image(image_path)
```
